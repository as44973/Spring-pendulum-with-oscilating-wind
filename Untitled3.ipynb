```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Parameters
l = 10
k = 12
m = 4
g = 9.81
beta = 0.3
A = 25  # Amplitude of wind force
omega_wind = 1.0  # Frequency of wind force
phi = np.pi / 2  # Phase of wind force
tmax = 250

# Equations of motion
def equations_of_motion(t, state):
    x, x_dot, y, y_dot = state
    
    wind_force_x = -A * omega_wind**2 * np.cos(omega_wind * t)*0
    wind_force_y = -A * omega_wind**2 * np.sin(omega_wind * t)
    
    x_double_dot = (-k * (1 - l / np.sqrt(x**2 + y**2)) * x - beta * (x_dot - wind_force_x)) / m*0
    y_double_dot = (-k * (1 - l / np.sqrt(x**2 + y**2)) * y - g*m - beta * (y_dot - wind_force_y)) / m
    
    return [x_dot, x_double_dot, y_dot, y_double_dot]

# Initial conditions [x, x_dot, y, y_dot]
initial_conditions = [1.0, 0.0, 1.0, 0.0]

sol = solve_ivp(equations_of_motion, [0, tmax], initial_conditions, t_eval=np.linspace(0, tmax, 1000))

x = sol.y[0]
y = sol.y[2]

plt.figure(figsize=(8, 6))
plt.plot(sol.t, y)
plt.xlabel('$t$')
plt.ylabel('$y$')
plt.title('Gibanje vzmetnega nihala z zranim uporom in periodino moduliranim vetrom')
plt.grid()

plt.show()

'''plt.figure(figsize=(8, 6))
plt.plot(y, tmax)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Spring Pendulum Motion with Air Resistance and Periodically Modulated Wind')
plt.grid()
plt.savefig("prva slika")

plt.show()'''
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Parameters
l = 1
k = 20
m = 40
g = 9.81
beta = 0.4
A = 5  # Amplitude of wind force
omega_wind = 1.0  # Frequency of wind force
phi = np.pi / 2  # Phase of wind force
tmax = 500

# Equations of motion
def equations_of_motion(t, state):
    x, x_dot, y, y_dot = state
    
    wind_force_x = -A * omega_wind**2 * np.cos(omega_wind * t)
    wind_force_y = -A * omega_wind**2 * np.cos(omega_wind * t)
    
    x_double_dot = (-k * (1 - l / np.sqrt(x**2 + y**2)) * x - beta * (x_dot - wind_force_x)) / m
    y_double_dot = (-k * (1 - l / np.sqrt(x**2 + y**2)) * y - m * g - beta * (y_dot - wind_force_y)) / m
    
    return [x_dot, x_double_dot, y_dot, y_double_dot]

# Generate initial conditions
num_initial_conditions = 200
initial_conditions = np.random.rand(num_initial_conditions, 4) * [100, 100, 100, 100]

# Function to calculate derivatives for plotting
def derivatives(t, state):
    x, x_dot, y, y_dot = state
    x_double_dot = (-k * (1 - l / np.sqrt(x**2 + y**2)) * x - beta * x_dot) / m
    y_double_dot = (-k * (1 - l / np.sqrt(x**2 + y**2)) * y - m * g - beta * y_dot) / m
    return [x_dot, x_double_dot, y_dot, y_double_dot]

# Plot phase space for x and x_dot
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
for initial_condition in initial_conditions:
    sol = solve_ivp(derivatives, [0, tmax], initial_condition, t_eval=np.linspace(0, tmax, 1000), method='RK45')
    plt.plot(sol.y[0], sol.y[1])

plt.xlabel("$x$")
plt.ylabel("$\\dot{x}$")
plt.title("Phase Space: $\\dot{x}$ vs. $x$")
plt.grid()

# Plot phase space for y and y_dot
plt.subplot(1, 2, 2)
for initial_condition in initial_conditions:
    sol = solve_ivp(derivatives, [0, tmax], initial_condition, t_eval=np.linspace(0, tmax, 1000), method='RK45')
    plt.plot(sol.y[2], sol.y[3])

plt.xlabel("$y$")
plt.ylabel("$\\dot{y}$")
plt.title("Phase Space: $\\dot{y}$ vs. $y$")
plt.grid()

plt.tight_layout()
plt.show()


# Function to find Poincar points
def find_poincare_points(sol, threshold):
    poincare_y = []
    poincare_y_dot = []

    for i in range(1, len(sol.t)):
        if (sol.y[0][i] >= threshold and sol.y[0][i - 1] < threshold) or (sol.y[0][i] <= threshold and sol.y[0][i - 1] > threshold):
            poincare_y.append(sol.y[2][i])
            poincare_y_dot.append(sol.y[3][i])

    return poincare_y, poincare_y_dot

# Iterate over different initial conditions
for i, initial_condition in enumerate(initial_conditions):
    sol = solve_ivp(equations_of_motion, [0, tmax], initial_condition, t_eval=np.linspace(0, tmax, 1000))
    
    poincare_y, poincare_y_dot = find_poincare_points(sol, 0)
    
    plt.scatter(poincare_y, poincare_y_dot, s=1, c='b', alpha=0.5)

plt.xlabel('$y$')
plt.ylabel("$\\dot{y}$")
plt.title("Poincar Section for $x = 0$")
plt.grid()
plt.show()


def find_poincare_points(sol, threshold):
    poincare_x = []
    poincare_x_dot = []

    for i in range(1, len(sol.t)):
        if (sol.y[2][i] >= threshold and sol.y[2][i - 1] < threshold) or (sol.y[2][i] <= threshold and sol.y[2][i - 1] > threshold):
            poincare_x.append(sol.y[0][i])
            poincare_x_dot.append(sol.y[1][i])

    return poincare_x, poincare_x_dot

# Iterate over different initial conditions
plt.figure(figsize=(6, 6))
for i, initial_condition in enumerate(initial_conditions):
    sol = solve_ivp(equations_of_motion, [0, tmax], initial_condition, t_eval=np.linspace(0, tmax, 1000))
    
    poincare_x, poincare_x_dot = find_poincare_points(sol, 0)
    
    plt.scatter(poincare_x, poincare_x_dot, s=1, c='b', alpha=0.5)

plt.xlabel('$x$')
plt.ylabel("$\\dot{x}$")
plt.title("Poincar Section for $y = 0$")
plt.grid()
plt.show()
```
