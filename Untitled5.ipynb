```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Parameters (adjust these values as needed)
l = 1  # Natural length of the spring
k = 10  # Spring constant
m = 10  # Mass of the pendulum
g = 9.81  # Gravitational acceleration
beta = 0.3  # Damping coefficient
A = 1  # Amplitude of wind force
omega_wind = 1.0  # Frequency of wind force
tmax =100  # Maximum time for simulation

# Derived parameters
omega_0 = np.sqrt(k / m)  # Natural frequency of the system
gamma = beta / (m * omega_0)  # Damping ratio (dimensionless)
A_tilde = A / (omega_0 * l)  # Dimensionless amplitude of the wind
g_tilde = g / (omega_0**2 * l)  # Dimensionless gravitational constant
omega_wind_tilde = omega_wind / omega_0  # Dimensionless wind frequency

# Equations of motion in dimensionless form
def dimensionless_equations(t, state):
    x_tilde, x_dot_tilde, y_tilde, y_dot_tilde = state
    
    # Distance of the mass from the spring's attachment point (dimensionless)
    r_tilde = np.sqrt(x_tilde**2 + y_tilde**2)
    
    # Restoring force of the spring (dimensionless)
    if r_tilde != 0:
        F_spring_x = - (x_tilde / r_tilde) * (r_tilde - 1)
        F_spring_y = - (y_tilde / r_tilde) * (r_tilde - 1)
    else:
        F_spring_x = 0
        F_spring_y = 0
    
    # Damping force proportional to velocity (dimensionless)
    F_damping_x = -gamma * (x_dot_tilde - A_tilde * np.cos(omega_wind_tilde * t))
    F_damping_y = -gamma * (y_dot_tilde - A_tilde * np.sin(omega_wind_tilde * t))
    
    # Equations of motion (dimensionless)
    x_double_dot_tilde = F_spring_x + F_damping_x
    y_double_dot_tilde = F_spring_y + F_damping_y - g_tilde
    
    return [x_dot_tilde, x_double_dot_tilde, y_dot_tilde, y_double_dot_tilde]

# Initial conditions [x_tilde, x_dot_tilde, y_tilde, y_dot_tilde]
initial_conditions = [1.0, 0.0, 0.0, 0.0]

# Solving the system of ODEs (dimensionless time and variables)
t_eval = np.linspace(0, omega_0 * tmax, 1000)
sol = solve_ivp(dimensionless_equations, [0, omega_0 * tmax], initial_conditions, t_eval=t_eval)

# Extracting the dimensionless x, y, velocity, and acceleration
x_tilde = sol.y[0]
x_dot_tilde = sol.y[1]
y_tilde = sol.y[2]
y_dot_tilde = sol.y[3]

# Time variable
t = t_eval / omega_0  # Converting to real time

# Calculate energies
r_tilde = np.sqrt(x_tilde**2 + y_tilde**2)

# Kinetic energy (dimensionless)
Ekin = 0.5 * (x_dot_tilde**2 + y_dot_tilde**2)

# Potential energy (dimensionless)
V_spring = 0.5 * (r_tilde - 1)**2
V_gravity = g_tilde * (1 + y_tilde)
V = V_spring + V_gravity

# Total energy
En = Ekin + V

# Plotting
plt.figure(figsize=(15, 10))

# Position and velocity plots
plt.subplot(3, 1, 1)
plt.plot(t, x_tilde, label=f'$\\tilde{{x}}$ (m={m}, k={k}, A={A})')
plt.plot(t, y_tilde, label=f'$\\tilde{{y}}$ (m={m}, k={k}, A={A})')
plt.xlabel('as(s)')
plt.ylabel('Pozicija')
plt.title('Pozicija v odvisnosti od asa t')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(t, x_dot_tilde, label=f'$\\dot{{\\tilde{{x}}}}$ (m={m}, k={k}, A={A})')
plt.plot(t, y_dot_tilde, label=f'$\\dot{{\\tilde{{y}}}}$ (m={m}, k={k}, A={A})')
plt.xlabel('Time (s)')
plt.ylabel('Hitrost')
plt.title('Hitrost v odvisnosti od asa')
plt.legend()


# Acceleration plot
accel_x = np.gradient(x_dot_tilde, t)
accel_y = np.gradient(y_dot_tilde, t)
plt.subplot(3, 1, 3)
plt.plot(t, accel_x, label=f'$\\ddot{{\\tilde{{x}}}}$ (m={m}, k={k}, A={A})')
plt.plot(t, accel_y, label=f'$\\ddot{{\\tilde{{y}}}}$ (m={m}, k={k}, A={A})')
plt.xlabel('Time (s)')
plt.ylabel('Pospeek')
plt.title('Pospeek v odvisnosti od asa')
plt.legend()
plt.savefig('pospeek 100 m')
plt.tight_layout()
plt.show()

# Energy plot
plt.figure(figsize=(8, 6))
plt.plot(t, Ekin, label=f'$E_{{kin}}$ (m={m}, k={k}, A={A})')
plt.plot(t, V, label=f'$V$ (m={m}, k={k}, A={A})')
plt.plot(t, En, label=f'$E_{{total}}$ (m={m}, k={k}, A={A})')
plt.xlabel('$t$')
plt.ylabel('$E$')
plt.title('Energija v odvisnosti od asa')
plt.legend()
plt.savefig('energija 100 m')
plt.tight_layout()
plt.show()

# Trajectory plot
plt.figure(figsize=(8, 6))
plt.plot(x_tilde, y_tilde, label=f'Trajektorija (m={m}, k={k}, A={A})')
plt.xlabel('$\\tilde{x}$')
plt.ylabel('$\\tilde{y}$')
plt.title('Trajektorija vzmetnega nihala')
plt.legend()
plt.savefig('trajektorija 100 m')
plt.show()

```
